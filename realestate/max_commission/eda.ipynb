```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hist.csv')
#df = df.loc[df['commission']!=0]
sorteddf = df.groupby(by=['rooms','floors','baths','lotSize','price']).sum().reset_index()
filtered_df = sorteddf.loc[(sorteddf['rooms'] == 1) & (sorteddf['floors'] == 2) & (sorteddf['baths'] == 2) ]
print(filtered_df.head())
```

```python
# Filter the DataFrame for rows where rooms == 1 and floors == 1
filtered_df = sorteddf.loc[(sorteddf['saleResult']== 'Sold')]
#filtered_df = sorteddf.loc[(sorteddf['rooms'] == 1) & (sorteddf['floors'] == 1) & (sorteddf['baths'] == 1) ]
sns.relplot(data=sorteddf, x='lotSize', y='price', kind='scatter', height=6, aspect=1.5,hue='saleResult',alpha=0.025,col='floors')
plt.show()
# Display the filtered DataFrame
# 
print(filtered_df)
```
